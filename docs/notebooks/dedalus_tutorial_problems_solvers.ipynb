## Problems

Systems of differential equations in Dedalus are represented in the form:

$$\mathcal{M} \cdot \partial_t \mathcal{X} + \mathcal{L} \cdot \mathcal{X} = \mathcal{F}$$

where $\mathcal{X}$ is a state-vector of fields, $\mathcal{F}$ is a set of generally nonlinear expressions represented by operator trees, and $\mathcal{M}$ and $\mathcal{L}$ are linear differential operators of the form $\mathcal{C} = \mathcal{C}_0 + \mathcal{C}_1 \partial_z$, where $\mathcal{C_0}$ and $\mathcal{C_1}$ are matrices of coefficients and derivatives along the first $(N-1)$ dimensions.

Dedalus includes a symbolic parser that takes equations specified in plain text, and manipulates them into the above matrix form.  The matrix form requires the equations to be first-order in $t$ and $z$ derivatives, and must only contain linear terms on the left-hand-side (the entire RHS is parsed as an operator tree).  

To create a `problem` object, you must specify names for the fields that appear in the equations.  Let's start setting up the KdV-Burgers equation:

$$\partial_t u + u \partial_x u = a \partial_{xx} u + b \partial_{xxx} u$$

The KdV-Burgers equation only has one primative field but is third-order, so we introduce two extra fields to reduce the equation to first-order.

First, we create a domain:

```python
x_basis = de.Fourier('x',1024, interval=(-2, 8), dealias=3/2)
domain = de.Domain([x_basis], np.float64)
```

and now we create the problem object, giving it the domain, and listing our variables. This problem is an Initial Value Problem, so we use the `IVP` problem provided by dedalus:

```python
problem = de.IVP(domain, 
                 variables=['u', 'ux', 'uxx'])
```

Then we can enter the first-order equation, and the equations defining the introduced variables, in plain text, defining parameters first:

```python
problem.parameters['a'] = 2e-4
problem.parameters['b'] = 1e-4

problem.add_equation("dt(u) - a*dx(ux) - b*dx(uxx) = -u*ux")
problem.add_equation("ux - dx(u) = 0")
problem.add_equation("uxx - dx(ux) = 0")
```

## Solvers

The solver classes are for solving differential types of differential equations.  There are currently solvers implemented for linear eigenvalue problems, linear boundary-value problems, and (general) initial-value problems.

To create a IVP solver, we pick a timestepper (several multistep and Runge-Kutta IMEX schemes are available) and hand it to the `.build_solver` method of the problem:

```python
solver = problem.build_solver(de.timesteppers.SBDF2)
```

### Stop criteria

Simulation stop criteria are specified by setting the `stop_teration`, `stop_wall_time` (seconds), and/or `stop_sim_time` attributes on the solver.  

Let's stop after 5000 iterations:

```python
# Stop stopping criteria
solver.stop_sim_time = np.inf
solver.stop_wall_time = np.inf
solver.stop_iteration = 5000
```

### State system

Attached to the solver is a system of fields representing the solution state-vector.  We can access the state fields using the names we specified when we created the problem, and then set their initial values:

```python
# Reference local grid and state fields
x = domain.grid(0)
u = solver.state['u']
ux = solver.state['ux']
uxx = solver.state['uxx']

# Smooth triangle with support in (-1, 1)
n = 20
u['g'] = np.log(1 + np.cosh(n)**2/np.cosh(n*x)**2) / (2*n)
u.differentiate(0, out=ux)
ux.differentiate(0, out=uxx)
```

### Time-stepping

The state system can be advanced by passing a step-size `dt` to the `solver.step` method.  

Using the `solver.ok` attribute, which will return `False` once a stopping criteria is reached, let's evolve the state, copying the grid values of `u` every few iterations.  This should take less than 10 seconds on most machines.

```python
# Setup storage
u_list = [np.copy(u['g'])]
t_list = [solver.sim_time]

# Main loop
dt = 2e-3
while solver.ok:
    solver.step(dt)
    if solver.iteration % 20 == 0:
        u_list.append(np.copy(u['g']))
        t_list.append(solver.sim_time)
```

Let's make a space-time plot of the solution on the full dealiased grid:

```python
from dedalus.extras import plot_tools

# Convert storage to arrays
u_array = np.array(u_list)
t_array = np.array(t_list)

x_da = domain.grid(0, scales=domain.dealias)

# Plot
xmesh, ymesh = plot_tools.quad_mesh(x=x_da, y=t_array)
plt.figure(figsize=(10,6))
plt.pcolormesh(xmesh, ymesh, u_array, cmap='RdBu_r')
plt.axis(plot_tools.pad_limits(xmesh, ymesh))
plt.colorbar()
plt.xlabel('x')
plt.ylabel('t')
plt.title('A dispersive shock!')
```
