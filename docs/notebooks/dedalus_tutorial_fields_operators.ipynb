# 2. Fields and Operators

This notebook covers the basics of setting up and interacting with field and operator objects in Dedalus.

First, we'll import the public interface:

```python
from dedalus2 import public as de
```

## Fields

Fields represent a scalar field defined over a domain.  A field can be directly instantiated from the `Field` class by passing a domain object, but it's better to create them using the `domain.new_field` method, which collects dereferenced fields for redistribution when possible.  Let's crank up the resolution (a little) and build a field:

```python
xbasis = de.Fourier(64, interval=(0,2), dealias=2/3)
ybasis = de.Chebyshev(64, interval=(0,1), dealias=2/3)
domain = de.Domain([xbasis, ybasis], grid_dtype=np.float64, mesh=[1])
f = domain.new_field()
f.name = 'f'
```

We also gave the field a name -- something which is automatically done for the state fields when solving a problem in Dedalus (we'll see more about this in the next notebook), but we just do manually, for now.

The current transform/distribution state of a field's data is described by a `layout` object, as discussed in the "Bases and Domains" notebook, and can be accessed as a field attribute:

```python
f.layout.grid_space
```

Field data can be written and retrieved in grid or coefficient space using `'g'` and `'c'` keys.  Let's assign our field some data using the local domain grids:

```python
x, y = domain.grids()
f['g'] = np.cos(2*2*np.pi*x) * np.exp(-2*y)
```

To plot it, we'll use some helper functions from the plot_tools extra:

```python
from dedalus_extras import plot_tools

def matrix_plot(matrix, clim):
    if np.isscalar(clim):
        clim = (-clim, clim)
    plt.imshow(matrix, clim=clim, cmap='RdBu_r')
    plt.colorbar()
    
def sparse_matrix_plot(matrix, clim):
    matrix_plot(matrix.todense(), clim)
    
def field_plot_2d(domain, field):
    # Plot and data axes are reversed
    y = domain.grid(0)
    x = domain.grid(1)
    xmesh, ymesh = plot_tools.quad_mesh(x.flatten(), y.flatten())
    plt.pcolormesh(xmesh, ymesh, field['g'], cmap='RdBu_r')
    plt.axis(plot_tools.pad_limits(x, y));
    plt.colorbar()
```

```python
field_plot_2d(domain, f)
```

And now let's take a look at the magnitude of the coefficients:

```python
matrix_plot(np.log10(np.abs(f['c'])), clim=(-20, 0))
```

In addition to the key interface, we can change the layout of a field using the `require_coeff_space`, `require_grid_space`, `require_layout`, and `require_local` methods.

## Operators

Mathematical operations involving fields are done through `Operator` objects.  Operators are classes representing various operations, such as arithmetic, differentiation, integration, and interpolation.  An instance of an operator class represents the *deferred evaluation* of that operation, performed with the supplied arguments.

Let's look at a simple operator using the field we've previously built:

```python
add_op = 10 + f
add_op
```

To actually compute the corresponding operation we use the `evaluate` method, which returns a new field with the result:

```python
add_out = add_op.evaluate()
field_plot_2d(domain, add_out)
```

Operator instances can be passed as arguments to other operators, building trees that represent more complicated expressions:

```python
tree_op = 2*add_op + f*f
tree_op
```

Reading these signatures can be a little cumbersome, but we can plot the operator's structure using a helper from dedalus2.tools:

```python
from dedalus2.tools.plot_op import plot_operator
plot_operator(tree_op, figsize=8, fontsize=12)
```

And evaluating it:

```python
tree_out = tree_op.evaluate()
field_plot_2d(domain, tree_out)
```

A key point is that the operator objects use deferred evaluation and so really represent a "symbolic" operation on the field arguments.  If we change the values of the field and re-evaluate the operator, we get a new result:

```python
f['g'] = np.sin(2*np.pi*x) * np.cos(4*np.pi*y)
field_plot_2d(domain, f)
```

```python
tree_out = tree_op.evaluate()
field_plot_2d(domain, tree_out)
```

In addition to arithmetic, operators are used differentiation, integration, interpolation, etc.

Derivative operators are available through the `domain.diff_ops` attribute, and can be explicitly evaluated into another field using the `field.differentiate` method:

```python
f_x_op = domain.diff_ops[0](f)
f_x_out = f_x_op.evaluate()

f_x = domain.new_field()
f.differentiate(0, out=f_x)

np.allclose(f_x['g'], f_x_out['g'])
```

```python
field_plot_2d(domain, f_x)
```

Interpolation and integration can be performed along one or several axes, and similarly can be used as operators or explicitly evaluated through field methods:

```python
# Evaluate total integral
f_int = f.integrate(xbasis, ybasis)
f_int['c'][0,0]
```

```python
# Interpolate along y axis
f_interp_op = de.operators.Interpolate(f, ybasis, 0.5)
f_interp = f_interp_op.evaluate()
plt.plot(domain.grid(0), f_interp['g'][:,0])
```
